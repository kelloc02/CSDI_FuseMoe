```python
import pickle
import matplotlib.pyplot as plt

foldername = "/playpen-shared/kechengli/workspace/Fusemoe/CSDI-main/save/ihm_fold0_20251118_021210/"
with open(foldername + "/loss_history.pk", "rb") as f:
    loss_hist = pickle.load(f)

train_loss = loss_hist["train"]                # list[float]
valid_loss = loss_hist["valid"]                # list[(epoch, loss)]

plt.plot(range(1, len(train_loss)+1), train_loss, label="train")

if valid_loss:
    epochs_v, vals_v = zip(*valid_loss)
    plt.plot(epochs_v, vals_v, label="valid")

plt.xlabel("epoch")
plt.ylabel("loss")
plt.legend()
plt.title("Training / Validation Loss Curve")
plt.show()
```
